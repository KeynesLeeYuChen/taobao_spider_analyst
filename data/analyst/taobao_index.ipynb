# 处理商品表格

```python
filename = 'taobao.csv'
colnames = ['title','price','deal','location','shop','url']
data = pd.read_csv(filename,names=colnames)
```

```python
data['price'] = data['price'].apply(lambda x : x[1:])
```

```python
data['price'] = data['price'].astype('float')
```

```python
data['profit'] = data['price']*data['deal']
```

```python
data['tmall'] = data['url'].str.contains('tmall')
```

```python
data_drop = data.drop_duplicates('title')
```

```python
data_drop['tmall'] = data_drop['tmall'].astype(str).apply(lambda x : x.replace('True','天猫'))
```

```python
data_drop['tmall'] = data_drop['tmall'].astype(str).apply(lambda x : x.replace('False','其他'))
```

```python
data_drop = data_drop.reset_index()
```

```python
data_drop['mall'] = data_drop['shop'].str.contains('店')
```

```python
edit_index = data_drop[(data_drop['tmall'].str.contains('天猫'))&(data_drop['mall']==False)]['mall'].index
```

```python
for i in edit_index:
    data_drop['mall'][i] = True
```

```python
data_drop['mall'] = data_drop['mall'].astype(str).apply(lambda x : x.replace('True','商家'))
```

```python
data_drop['mall'] = data_drop['mall'].astype(str).apply(lambda x : x.replace('False','个人'))
```

```python
data_drop['brand'] = data_drop['title']
```

```python
def brandName(s):
    if s.find('兰蔻') != -1:
        return "兰蔻"
    elif (s.find('MAC') != -1)| (s.find('魅可') != -1):
        return "MAC/魅可"
    elif (s.find('ysl') != -1) | (s.find('圣罗兰') != -1):
        return "ysl/圣罗兰"
    elif (s.find('Dior') != -1) | (s.find('迪奥') != -1):
        return "Dior/迪奥"
    elif s.find('纪梵希') != -1:
        return "Givenchy/纪梵希"
    elif (s.find('香奈儿') != -1) | (s.find('香奈尔') != -1) | (s.find('Chanel') != -1):
        return "Chanel/香奈儿"
    elif (s.find('美宝莲') != -1) | (s.find('Maybelline') != -1):
        return "Maybelline/美宝莲"
    elif (s.find('Chioture') != -1) | (s.find('稚优泉') != -1):
        return "Chioture/稚优泉"
    elif (s.find('Ford') != -1) | (s.find('福特') != -1):
        return "Ford/福特"
    elif (s.find('The Saem') != -1) | (s.find('得群') != -1):
        return "The Saem/得群"
    else:
        return "其他"  
```

```python
data_drop['brand'] = data_drop['brand'].apply(brandName)
```

```python
data_drop['pricelevel'] = data_drop['price']
```

```python
def priceLevel(s):
    if (s>=0.0) & (s<20.0):
        return '0-20'
    
    elif (s>=20.0) & (s<40.0):
        return '20-40'
    
    elif (s>=40.0) & (s<60.0):
        return '40-60'
    
    elif (s>=60.0) & (s<80.0):
        return '60-80'
    
    elif (s>=80.0) & (s<100.0):
        return '80-100'
    
    elif (s>=100.0) & (s<120.0):
        return '100-120'
    
    elif (s>=120.0) & (s<140.0):
        return '120-140'
    
    elif (s>=140.0) & (s<160.0):
        return '140-160'
    
    elif (s>=160.0) & (s<180.0):
        return '160-180'
    
    elif (s>=180.0) & (s<200.0):
        return '180-200'
    
    elif (s>=200.0) & (s<220.0):
        return '200-220'
    
    elif (s>=220.0) & (s<240.0):
        return '220-240'
    
    elif (s>=240.0) & (s<260.0):
        return '240-260'
    
    elif (s>=260.0) & (s<280.0):
        return '260-280'
    
    elif (s>=280.0) & (s<300.0):
        return '280-300'
    
    elif (s>=300.0) & (s<320.0):
        return '300-320'
    
    elif (s>=320.0) & (s<340.0):
        return '320-340'
    
    elif (s>=340.0) & (s<360.0):
        return '340-360'
    
    elif (s>=360.0) & (s<380.0):
        return '360-380'
    
    elif (s>=380.0) & (s<400.0):
        return '380-400'
    
    elif (s>400.0):
        return '400以上'
```

```python
data_drop['pricelevel'] = data_drop['pricelevel'].apply(priceLevel)
```

```python
def locationProvince(s):
    if s.find(' ') != -1:
        return s[:s.find(' ')]
    else:
        return s

def locationCity(s):
    if s.find(' ') != -1:
        return s[s.find(' '):]
    else:
        return s
```

```python
data_drop['province'] = data_drop['location'].apply(locationProvince)
```

```python
data_drop['city'] = data_drop['location'].apply(locationCity)
```

```python
data_drop[(data_drop['title'].str.contains('礼')) & (data_drop['price']>400)] = None
```

```python
data_drop = data_drop.dropna()
```

```python
data_drop.to_csv('taobao_drop_map.csv')
```
